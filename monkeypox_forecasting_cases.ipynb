```python
def symp_dict(row):
    symp_row=row
    symp_list=symp_row.split(",")
    if symp_row=="Unknown":
        score=0
    else:
        score=len(symp_list)
    return score
```

```python
df['Symptoms'].fillna("Unknown",inplace=True)
```

```python
df['Symp_count']=df['Symptoms'].apply(symp_dict)
```
